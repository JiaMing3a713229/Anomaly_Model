```python
padding_pos = [[0, 19],[0, 19],[0, 19],[1, 18],[8, 17],[10, 16]]
insole_arr = np.zeros((19,6))
print(insole_arr.shape)
```

```python
def insole_padding(padding_data):
    padding_pos = [[0, 19],[0, 19],[0, 19],[1, 18],[8, 17],[10, 16]]
    insole_arr = np.zeros((19,6))
    row = 0
    column = 0
    i = 0
    for start_pos, end_pos in padding_pos:
        for k in range(start_pos, end_pos):
            insole_arr[18 - k][5 - column] = padding_data[i]
            i += 1
        column += 1
    return insole_arr
        
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
from sklearn.neighbors import KernelDensity
```

### 模型資料前處理

```python
# 将数据转换为适当的形状 (35350, 12, 20, 1)，添加通道维度
insole_traindata = datasets.reshape(insole_len, 19, 6, 1)/255
print(insole_traindata.shape)
```

#### Model Articture

```python
# 构建卷积自编码器
model = models.Sequential()

# 编码器部分
model.add(layers.Conv2D(32, (3, 2), activation='relu', padding='same', input_shape=(19, 6, 1)))
model.add(layers.Conv2D(64, (3, 2), activation='relu', padding='same'))

# 解码器部分
model.add(layers.Conv2DTranspose(64, (3, 2), activation='relu', padding='same'))
model.add(layers.Conv2DTranspose(32, (3, 2), activation='relu', padding='same'))

# 输出层
model.add(layers.Conv2D(1, (3, 2), activation='sigmoid', padding='same'))


model.summary()
```

#### Train Model

```python
learning_rate = 0.0001  # You can adjust this value
optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)

model.compile(optimizer=optimizer, loss='mse', metrics=['mse'])

# Train model
history = model.fit(insole_traindata, insole_traindata, epochs=1000, batch_size=1000, shuffle=True)
```

#### Test dataset

```python
insole_anomalydata_np = np.array(anomaly_datasets)  
insole_anomalydata = insole_anomalydata_np.reshape(insole_anomalydata_np.shape[0], 19, 6, 1)/255
print(insole_anomalydata.shape)
```

```python
# 使用训练好的模型进行预测
decoded_samples = model.predict(insole_traindata)
decoded_samples1 = model.predict(insole_anomalydata)

# 计算重构误差
mse = tf.keras.losses.MeanSquaredError()
reconstruction_error = mse(insole_traindata, decoded_samples).numpy()
print(f"Reconstruction Error: {reconstruction_error * 10000}")

# 计算重构误差
mse1 = tf.keras.losses.MeanSquaredError()
reconstruction_error1 = mse1(insole_anomalydata, decoded_samples1).numpy()
print(f"Reconstruction Error: {reconstruction_error1 * 10000}")
```

```python
loss = history.history['loss']

# 绘制损失曲线
plt.plot(loss)
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.show()
```
